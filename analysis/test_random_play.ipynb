```python
import logging
import random
from time import sleep
import unittest
import sys

sys.path.append('.')
from Simulation import Simulation
from agents.Agent import RandomAgent, SJAgent
from env.Actions import DeclareAction, DontDeclareAction, PlaceKittyAction
from env.utils import AbsolutePosition, Declaration, TrumpSuite
from env.CardSet import CardSet
from env.Game import Game

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

# check distributions of random games

```python
r = []
for i in range(0, 500):
    random.seed(i) # DEBUG: seed=2 is problematic
    sim = Simulation(RandomAgent('NS'), RandomAgent('WE'), enable_chaodi=False)
    # logging.getLogger().setLevel(logging.DEBUG)
    while sim.step()[0]: 
        pass
    info, rounds, leaders, moves = sim.game_engine.return_status()
    r.append(info['opponents_total_points'])
```

```python
plt.plot(np.arange(500), r)
plt.xlabel('game id')
plt.ylabel('opponents total points')
plt.show()
```

```python
plt.hist(r, bins=np.linspace(0,200,15), histtype='step')
plt.xlabel('opponents total points')
plt.axvline(80,linestyle='--',color='black',label='80')
plt.legend()
plt.show()
```
