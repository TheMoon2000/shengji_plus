```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
from typing import List
```

```python
def graph_win_rate(model_names: List[str], save=True):

    plt.figure(facecolor='white', figsize=(9, 5))
    plt.title(f'{len(model_names)} models learning progress (against random model)')
    plt.xlabel('Games Played')
    plt.ylabel('Win rate')
    for name in model_names:
        with open(f'../{name}/stats.pkl', 'rb') as f:
            data = pickle.load(f)
        iterations = [x['iterations'] for x in data]
        win_rates = [x['win_counts'] for x in data]
        plt.plot(iterations, win_rates, label=name, alpha=0.7)
        plt.legend()
    if len(model_names) == 1:
        plt.savefig(f'../{model_names[0]}/win_rate.png')

def graph_apd(model_names: List[str], save=True):
    plt.figure(facecolor='white', figsize=(9, 5))
    plt.title(f'{len(model_names)} models average point difference (against random model)')
    plt.xlabel('Games Played')
    plt.ylabel('Average Point Difference')
    for name in model_names:
        with open(f'../{name}/stats.pkl', 'rb') as f:
            data = pickle.load(f)
        iterations = [x['iterations'] for x in data]
        apd = [x['avg_points'][0] - x['avg_points'][1] for x in data]
        plt.plot(iterations, apd, label=name, alpha=0.7)
    plt.legend()

    if len(model_names) == 1:
        plt.savefig(f'../{model_names[0]}/apd.png')

def graph_level_rate(model_names: List[str], save=True):
    plt.figure(facecolor='white', figsize=(9, 5))
    plt.title(f'{len(model_names)} models level rate (against random model)')
    plt.xlabel('Games Played')
    plt.ylabel('Level rate')
    for name in model_names:
        with open(f'../{name}/stats.pkl', 'rb') as f:
            data = pickle.load(f)
        iterations = [x['iterations'] for x in data]
        level_rates = [x['level_counts'] for x in data]
        plt.plot(iterations, level_rates, label=name, alpha=0.7)
        plt.legend()
    if len(model_names) == 1:
        plt.savefig(f'../{model_names[0]}/level_rate.png')
```

```python
graph_win_rate('small')
```

```python
graph_win_rate(['small_fix_batches'])
graph_apd(['small_fix_batches'])
```

```python
graph_win_rate(['small_discount=0.99', 'small_real_0.7', 'small_discount=0.6', 'small_hash_20bit', 'small_e=0.05'])
graph_apd(['small_discount=0.99',  'small_real_0.7', 'small_discount=0.6', 'small_e=0.05'])
```
