```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
from typing import List
```

```python
def graph_win_rate(model_names: List[str], save=True):

    plt.figure(facecolor='white', figsize=(8, 5))
    plt.title(f'{model_names} learning progress (against random model)')
    plt.xlabel('Games Played')
    plt.ylabel('Win rate')
    for name in model_names:
        with open(f'../{name}/stats.pkl', 'rb') as f:
            data = pickle.load(f)
        iterations = [x['iterations'] for x in data]
        win_rates = [x['win_counts'] for x in data]
        plt.plot(iterations, win_rates, label=name, alpha=0.8)
        plt.legend()
    if len(model_names) == 1:
        plt.savefig(f'../{model_names[0]}/win_rate.png')

def graph_apd(model_names: List[str], save=True):
    plt.figure(facecolor='white', figsize=(8, 5))
    plt.title(f'{model_names} learning progress (against random model)')
    plt.xlabel('Games Played')
    plt.ylabel('Average Point Difference')
    for name in model_names:
        with open(f'../{name}/stats.pkl', 'rb') as f:
            data = pickle.load(f)
        iterations = [x['iterations'] for x in data]
        apd = [x['avg_points'][0] - x['avg_points'][1] for x in data]
        plt.plot(iterations, apd, label=name)
    plt.legend()

    if len(model_names) == 1:
        plt.savefig(f'../{model_names[0]}/apd.png')
```

```python
graph_win_rate('small')
```

```python
graph_win_rate('small_epsilon_decay_1.25')
graph_apd('small_epsilon_decay_1.25')
```

```python
graph_win_rate('small_epsilon_decay_1.2')
graph_apd('small_epsilon_decay_1.2')
```

```python
graph_win_rate('small_parallel')
graph_apd('small_parallel')
```

```python
graph_win_rate(['small_fix_batches'])
graph_apd(['small_fix_batches'])
```

```python
graph_win_rate(['small_4layer_main'])
graph_apd(['small_4layer_main'])
```
