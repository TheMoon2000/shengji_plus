```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
def graph_win_rate(model_name: str, save=True):
    with open(f'../{model_name}/stats.pkl', 'rb') as f:
        data = pickle.load(f)
    iterations = [x['iterations'] for x in data]
    win_rates = [x['win_counts'] for x in data]

    plt.figure(facecolor='white', figsize=(8, 5))
    plt.title(f'{model_name} learning progress (against random model)')
    plt.xlabel('Games Played')
    plt.ylabel('Win rate')
    plt.plot(iterations, win_rates)
    if save:
        plt.savefig(f'../{model_name}/win_rate.png')

def graph_apd(model_name: str, save=True):
    with open(f'../{model_name}/stats.pkl', 'rb') as f:
        data = pickle.load(f)
    iterations = [x['iterations'] for x in data]
    apd = [x['avg_points'][0] - x['avg_points'][1] for x in data]

    plt.figure(facecolor='white', figsize=(8, 5))
    plt.title(f'{model_name} learning progress (against random model)')
    plt.xlabel('Games Played')
    plt.ylabel('Average Point Difference')
    plt.plot(iterations, apd)
    if save:
        plt.savefig(f'../{model_name}/apd.png')
```

```python
graph_win_rate('small')
```

```python
graph_win_rate('small_explore')
graph_apd('small_explore')
```
